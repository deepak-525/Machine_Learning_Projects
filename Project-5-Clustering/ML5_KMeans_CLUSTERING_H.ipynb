## import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")
```

## load data

```python
pen_df=pd.read_csv(r"C:\DATASCIENCE\dataset\penguins.csv")
```

```python
pen_df.head()
```

## Removing outliers

```python
pen_df.boxplot()
```

```python
# Drop rows with missing values
pen_df = pen_df.dropna()
```

## Removing Outliers using Interquartile Range (IQR) Method

```python
# List of numeric columns
numeric_columns = ['culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']

for column in numeric_columns:
    Q1 = pen_df[column].quantile(0.25)
    Q3 = pen_df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    pen_df = pen_df[(pen_df[column] >= lower_bound) & (pen_df[column] <= upper_bound)]
```

```python
# Check the first few rows of the cleaned dataset


# Boxplot again to verify
pen_df[['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']].boxplot()
plt.show()
print(pen_df.head())
```

## one hot encoding

```python
df = pd.get_dummies(pen_df).drop("sex_.", axis=1)
df.head()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(df)
penguins_preprocessed = pd.DataFrame(data=X, columns=df.columns)
penguins_preprocessed.head(10)
```

## Principal Component Analysis (PCA) for Dimensionality Reduction

```python
pca = PCA(n_components=None)
dfx_pca = pca.fit(penguins_preprocessed)
dfx_pca.explained_variance_ratio_
n_components = sum(dfx_pca.explained_variance_ratio_ > 0.1)
pca = PCA(n_components=n_components)
penguins_PCA = pca.fit_transform(penguins_preprocessed)
print(n_components)
```

## Determining Optimal Number of Clusters using Elbow Method

```python
inertia = []                           #wcss
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(penguins_PCA)
    inertia.append(kmeans.inertia_)
plt.plot(range(1, 10), inertia, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
n_clusters = 4
```

## Visualizing K-means Clustering Results

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(penguins_PCA)
plt.scatter(penguins_PCA[:, 0], penguins_PCA[:, 1], c=kmeans.labels_, cmap="viridis")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title(f"K-means Clustering (K={n_clusters})")
plt.show()
```

## silhouette score

```python
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(penguins_PCA)
silhouette = silhouette_score(penguins_PCA, kmeans.labels_)
print(silhouette)
```
