## converting to list

```python
transactions = np.array(df.values, dtype=str).tolist()
```

## Generating Association Rules using Apriori Algorithm

```python
from apyori import apriori

rules = apriori(transactions, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
```

## Extracting Association Rules from Apriori Algorithm

```python
result = list(rules)

# rules = [(itemset1, (support, confidence, lift)), (itemset2, (support, confidence, lift))]

# result = [
#     (itemset1, (support, confidence, lift)),
#     (itemset2, (support, confidence, lift)),
#     ...
# ]

```
